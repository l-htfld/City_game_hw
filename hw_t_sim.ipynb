```python
import random
```

```python
class City:
    def __init__(self, name, budget, happy_lvl=50, eco_lvl=50, edu_lvl=50, migr_lvl=50, drug_lvl=50, gun_lvl=50, anarchy_lvl=50) -> None:
        self.name = name
        self.budget = budget  # твой бюджет города
        self.happy_lvl = happy_lvl #уровень счастья
        self.eco_lvl = eco_lvl #уровень экологии
        self.edu_lvl = edu_lvl #уровень образования
        self.migr_lvl = migr_lvl #контроль миграции
        self.drug_lvl = drug_lvl #оборот наркотиков
        self.gun_lvl = gun_lvl #свобода владения оружием
        self.anarchy_lvl = anarchy_lvl #уровень анархии

    def greeting(self):
        return('Приветствую, {}! Добро Пожаловать в симулятор города, где ты побудешь в роли мэра. Цель игры - построить утопию (довести все параматеры города до 100). \n Но сначала давай выберем идеологию! \n Выбери пожалуйста подходящую для тебя идеологию, в зависимости от уровня твоего безумия: Капитализм, Социализм, Анархия'.format(self.name))
    
    def choose_ideology(self, ideology):
        """Задаёт уровни параметров города в зависимости от выбранной идеологии"""
        if ideology.lower() == 'капитализм':
            self.happy_lvl = 70
            self.eco_lvl = 40
            self.edu_lvl = 60
            self.migr_lvl = 20
            self.drug_lvl = 30
            self.gun_lvl = 80
            self.anarchy_lvl = 10
            print("Ты выбрал Капитализм! Твой город теперь настроен по принципу: 'Капитализм, счастье и так далее'!")

        elif ideology.lower() == 'социализм':
            self.happy_lvl = 85
            self.eco_lvl = 70
            self.edu_lvl = 80
            self.migr_lvl = 60
            self.drug_lvl = 10
            self.gun_lvl = 20
            self.anarchy_lvl = 5
            print("Ты выбрал Социализм! Пойдешь иным путем!")

        elif ideology.lower() == 'анархия':
            self.happy_lvl = 50
            self.eco_lvl = 30
            self.edu_lvl = 20
            self.migr_lvl = 80
            self.drug_lvl = 90
            self.gun_lvl = 90
            self.anarchy_lvl = 100
            print("Ты выбрал Анархию! Начинай стрелять по окнам и выносить телевизоры!")

        else:
            print("Неизвестная идеология. Пожалуйста, выбери: Капитализм, Социализм или Анархия.")
    
    def display_city_features(self):
        """Выводит параметры города"""
        print(f"\nТекущие параметры твоего города ({self.name}):")
        print(f"1. Уровень счастья: {self.happy_lvl}")
        print(f"2. Уровень экологии: {self.eco_lvl}")
        print(f"3. Уровень образования: {self.edu_lvl}")
        print(f"4. Контроль миграции: {self.migr_lvl}")
        print(f"5. Оборот наркотиков: {self.drug_lvl}")
        print(f"6. Свобода владения оружием: {self.gun_lvl}")
        print(f"7. Уровень анархии: {self.anarchy_lvl}")
        print(f"Бюджет: {self.budget}")
        
    def check_utopia(self):
        """Проверяет, достиг ли город утопии (все параметры на 100)"""
        if (self.happy_lvl == 100 and 
            self.eco_lvl == 100 and 
            self.edu_lvl == 100 and 
            self.migr_lvl == 100 and 
            self.drug_lvl == 100 and 
            self.gun_lvl == 100 and 
            self.anarchy_lvl == 100):
            print("Поздравляем! Твой город достиг утопии!")
            return True
        else:
            return False
        
    def gun_shootout(self):
        """Внезапная перестрелка (шанс - 30%), если уровень оружия > 70"""
        if self.gun_lvl > 70:
            if random.random() < 0.3: 
                self.budget -= 10
                print("Произошла перестрелка в городе! Бюджет уменьшен на 10.")
                if self.budget < 0:
                    self.budget = 0
```

```python
class Building(City):
    def build_school(self):
        """Строительство школы увеличивает уровень образования, но расходует бюджет и снижает анархию"""
        if self.budget >= 10:
            self.edu_lvl += 10
            self.anarchy_lvl -= 10
            self.budget -= 10
            if self.edu_lvl > 100:
                self.edu_lvl = 100
            if self.anarchy_lvl < 100:
                self.anarchy_lvl = 100
            print(f"Построена школа! Уровень образования увеличен до {self.edu_lvl}, уровень анархии снижен до {self.anarchy_lvl}. Бюджет: {self.budget}")
        else:
            print("Недостаточно бюджета для строительства школы.")

    def build_hospital(self):
        """Строительство больницы увеличивает уровень счастья и экологии, но расходует бюджет"""
        if self.budget >= 15:
            self.happy_lvl += 10
            self.eco_lvl += 5
            self.budget -= 15
            if self.happy_lvl > 100:
                self.happy_lvl = 100
            if self.eco_lvl > 100:
                self.eco_lvl = 100
            print(f"Построена больница! Уровень счастья увеличен до {self.happy_lvl}, уровень экологии до {self.eco_lvl}. Бюджет: {self.budget}")
        else:
            print("Недостаточно бюджета для строительства больницы.")

    def build_factory(self):
        """Строительство завода уменьшает уровень экологии, но пополняет бюджет"""
        self.eco_lvl -= 15
        self.happy_lvl -= 5
        self.budget += 20
        if self.eco_lvl < 0:
            self.eco_lvl = 0
        if self.happy_lvl < 0:
            self.happy_lvl = 0
        print(f"Построен завод! Уровень экологии уменьшен до {self.eco_lvl}, уровень счастья до {self.happy_lvl}. Бюджет: {self.budget}")

    def build_den(self):
        """Строительство притона увеличивает бюджет и экологию"""
        self.budget += 10
        self.eco_lvl += 7
        if self.eco_lvl > 0:
            self.eco_lvl = 0
        print(f"Построен наркодистрибьютор! Экология увеличина до {self.eco_lvl}, бюджет увеличен до {self.budget}.")

    def enhance_migration(self):
        """Усиление миграционной политики увеличивает уровень счастья и экологии, но уменьшает бюджет"""
        if self.budget >= 5:
            self.budget -= 5
            self.happy_lvl += 9
            self.eco_lvl += 9
            if self.happy_lvl > 100:
                self.happy_lvl = 100
            if self.eco_lvl > 100:
                self.eco_lvl = 100
            print(f"Миграционная политика усилена! Уровень счастья увеличен до {self.happy_lvl}, экологии до {self.eco_lvl}. Бюджет: {self.budget}")
        else:
            print("Недостаточно бюджета для усиления миграционной политики.")

    def build_gun_shop(self):
        """Строительство оружейной увеличивает бюджет, уровень оружия и анархии"""
        self.budget += 10
        self.gun_lvl += 20
        self.anarchy_lvl += 25
        if self.gun_lvl > 100:
            self.gun_lvl = 100
        if self.anarchy_lvl > 100:
            self.anarchy_lvl = 100
        print(f"Построена оружейная! Уровень владения оружием увеличен до {self.gun_lvl}, уровень анархии до {self.anarchy_lvl}. Бюджет: {self.budget}")
```

```python
# Образец игры
city = Building('Player', budget=50)
print(city.greeting())
ideology = input("Введите идеологию (Капитализм, Социализм, Анархия): ")
city.choose_ideology(ideology)

while not city.check_utopia():
    city.display_city_features()
    build = input("\nВведи 'стоп', если хочешь завершить. \nЧто ты хочешь построить? (школа/больница/завод/притон/миграция/оружейная): ").lower()

    if build == "школа":
        city.build_school()
    elif build == "больница":
        city.build_hospital()
    elif build == "завод":
        city.build_factory()
    elif build == "притон":
        city.build_den()
    elif build == "миграция":
        city.enhance_migration()
    elif build == "оружейная":
        city.build_gun_shop()
    elif build == "стоп":
        break
    else:
        print("Неизвестное действие. Попробуй снова.")
        
    city.gun_shootout()
    
    if city.budget <= 0:
        print("Твой бюджет закончился. Игра окончена.")
        break

# Вывод финального состояния города
city.display_city_features()
```
